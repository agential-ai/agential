```python
question = 'Which rock band chose its name by drawing it out of a hat, Switchfoot or Midnight Oil?'

agent = SelfRefine(
    llm=llm,
    benchmark="hotpotqa",
    # kwargs.
    patience=2
)
```

```python
out = agent.generate(
    question=question,
    examples=HOTPOTQA_FEWSHOT_EXAMPLES_COT,  # HOTPOTQA_FEWSHOT_EXAMPLES_DIRECT, HOTPOTQA_FEWSHOT_EXAMPLES_REACT
    prompt=SELF_REFINE_INSTRUCTION_HOTPOTQA,
    critique_examples=HOTPOTQA_CRITIQUE_FEWSHOT_EXAMPLES,
    critique_prompt=SELF_REFINE_CRITIQUE_INSTRUCTION_HOTPOTQA,
    refine_examples=HOTPOTQA_REFINE_FEWSHOT_EXAMPLES,
    refine_prompt=SELF_REFINE_REFINE_INSTRUCTION_HOTPOTQA,
    additional_keys={},
    critique_additional_keys={},
    refine_additional_keys={},
    max_interactions=3,
    reset=True
)
```

```python
for step in out.additional_info:
    print("INPUT-----------------------")
    print(step.answer + "\n")
    print("OUTPUT-----------------------")
    print(step.critique + "\n")
    print("Critique INPUT-----------------------")
    print(step.critique_response.input_text + "\n")
    print("Critique OUTPUT-----------------------")
    print(step.critique_response.output_text + "\n")
```
