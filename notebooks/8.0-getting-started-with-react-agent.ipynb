Large Language Models (LLM) are a type of AI/ML model that are designed to interpret and generate text. They are used in various applications for websites and online tools, such as chatbots and virtual assistants, or even language translation and improving search engine responses.

The term "large" refers to two things; the massive neural network architecture itself that is required for the LLM (possibly consisting of millions of parameters), and to the vast amount of training data used to train the LLM itself.

These models are being used for significant advancements in automation, augmentation and innovation. (Good so far?)

### 1.2 What is ReAct?
So far, language models are getting better at reasoning and acting. They have demonstrated remarkable reasoning abilities (chain-of-thought), such as excellent arithmetic and reasoning tasks, and have impressive results in action as well (action plan generation).

However, these two topics have generally been implmented separately within most LLMs. What if these two capacities were combined?

This is what ReAct achieves, which assists the model in improving its decision-making and human interpretability. 

As the acronym suggests, ReAct conducts the LLM to combine both actions into one: Reason + Act. The combination of these two processes has shown immense success in prompting bigger language models, and also successful in fine-tuning smaller ones. (could add more?)
