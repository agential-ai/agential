```python
file_path = 'artifacts/run-0v398r3v-glamorousriver46_eval:v0/glamorous-river-46_eval.table.json'
```

```python
import json
import json

# Open the file and load the JSON data
with open(file_path, 'r') as file:
    file_input = json.load(file)
```

```python
import pandas as pd
df = pd.DataFrame(file_input['data'])
df.columns = file_input['columns']
print(df)
```

```python
filtered_correct_df = df[df['EM'] == 1]
```

```python
filtered_correct_df
```

```python
filtered_incorrect_df = df[df['EM'] == 0]
filtered_incorrect_df
```

```python
filtered_empty_df = df[
    (df['predicted_answer'].isnull() | (df['predicted_answer'].str.strip() == ''))  # Condition for empty/null predicted_answer
    & (df['EM'] == 0)  # Condition for EM == 0
]
filtered_empty_df
```

```python
# Create a new condition for 'predicted_answer' that is NOT empty or null
non_empty_condition = ~(df['predicted_answer'].isnull() | (df['predicted_answer'].str.strip() == ''))

# Optionally, combine this with other conditions like EM == 0
filtered_non_empty_df = df[non_empty_condition & (df['EM'] == 0)]
filtered_non_empty_df
```
