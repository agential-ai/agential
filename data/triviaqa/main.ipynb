```python
from datasets import load_dataset

data = load_dataset("mandarjoshi/trivia_qa", "rc", split="validation")
```

```python
len(data)
```

```python
import random
import json
import pickle

save_file_name = 'triviaqa'
sample_size = 500
seed = 42

random.seed(seed)

indices = list(range(len(data)))
random.shuffle(indices)

with open(f"{save_file_name}_s{seed}_indices.pkl", 'wb') as f:
    pickle.dump(indices, f)

data = [data[i] for i in indices[:sample_size]]

with open(f"{save_file_name}_s{seed}_sample{sample_size}.json", 'w') as json_file:
    json.dump(data, json_file, indent=4)
```

```python
with open("triviaqa_s42_sample500.json", "r") as f:
    data = json.load(f)
```

```python
from collections import defaultdict
result = defaultdict(list)


for item in data:
    for key, value in item.items():
        result[key].append(value)

result = dict(result)
```
