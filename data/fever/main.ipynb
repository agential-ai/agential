```python
import random
import json
import pickle

file_name = 'paper_dev.jsonl'
save_file_name = 'paper_dev'
sample_size = 500
seed = 42

random.seed(seed)

data = []
with open(file_name, 'r', encoding='utf-8') as f:
    for line in f:
        d = json.loads(line)
        data.append(d)

indices = list(range(len(data)))
random.shuffle(indices)

with open(f"{save_file_name}_s{seed}_indices.pkl", 'wb') as f:
    pickle.dump(indices, f)

data = [data[i] for i in indices[:sample_size]]

with open(f"{save_file_name}_s{seed}_sample{sample_size}.json", 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
